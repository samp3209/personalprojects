```python
import numpy as np
import tensorflow as tf
import random
import pandas as pd

class QNetwork(tf.keras.Model):
    def __init__(self, num_actions, input_shape):
        super(QNetwork, self).__init__()
        self.dense1 = tf.keras.layers.Dense(32, activation='relu', input_shape=(input_shape,))
        self.dense2 = tf.keras.layers.Dense(num_actions, activation='linear')

    def call(self, state):
        # Reshape the input to 2D if it's 1D
        if len(state.shape) == 1:
            state = tf.expand_dims(state, axis=0)
        x = self.dense1(state)
        return self.dense2(x)

class TicTacToeAgent:
    def __init__(self, num_actions, epsilon=0.1, gamma=0.9, learning_rate=0.001):
        self.num_actions = num_actions
        self.epsilon = epsilon
        self.gamma = gamma
        self.q_network = QNetwork(num_actions, input_shape=9)
        self.optimizer = tf.keras.optimizers.Adam(learning_rate)

    def choose_action(self, state):
        if random.uniform(0, 1) < self.epsilon:
            return random.randint(0, self.num_actions - 1)
        else:
            q_values = self.q_network(state)
            return np.argmax(q_values)

    def train(self, state, action, reward, next_state, done):
        with tf.GradientTape() as tape:
            target = reward + self.gamma * np.max(self.q_network(next_state)) * (1 - done)
            q_values = self.q_network(state)
            action_one_hot = tf.one_hot(action, self.num_actions)
            selected_q_value = tf.reduce_sum(tf.multiply(q_values, action_one_hot))
            loss = tf.losses.mean_squared_error(target, selected_q_value)

        grads = tape.gradient(loss, self.q_network.trainable_variables)
        self.optimizer.apply_gradients(zip(grads, self.q_network.trainable_variables))

class TicTacToe:
    EMPTY_CELL = -1

    def __init__(self, agent=None):
        # Initialize the board with None (empty cells)
        self.board = [self.EMPTY_CELL] * 9
        # Current player (0 for Player 1, 1 for Player 2)
        self.current_player = 0
        # DataFrame to store moves and outcomes for each game
        self.games_data = pd.DataFrame(columns=['Moves', 'Outcome'])
        # Reference to the TicTacToeAgent
        self.agent = agent

    def reset(self):
        # Reset the board and player
        self.board = [self.EMPTY_CELL] * 9
        self.current_player = 0

    def print_board(self):
        # Print the current state of the board
        for i in range(3):
            row = self.board[i * 3: (i + 1) * 3]
            print("|".join([" " if cell == self.EMPTY_CELL else str(cell) for cell in row]))
            if i < 2:
                print("-----")

    def print_board_after_move(self):
        print("Current Board:")
        self.print_board()
        print("----------------")

    def make_move(self, position):
        # Make a move on the board
        if self.board[position] != self.EMPTY_CELL:
            print("Invalid move. Cell already occupied.")
            return False  # Move unsuccessful

        self.board[position] = self.current_player
        self.current_player = 1 - self.current_player  # Switch player
        return True  # Move successful

    def check_winner(self):
        # Check for a win condition after the 5th move
        if len([cell for cell in self.board if cell != self.EMPTY_CELL]) >= 5:
            # Check rows, columns, and diagonals
            for i in range(3):
                # Check rows
                if (
                        self.board[i * 3] == self.board[i * 3 + 1] == self.board[i * 3 + 2] != self.EMPTY_CELL
                ):
                    return True
                # Check columns
                if (
                        self.board[i] == self.board[i + 3] == self.board[i + 6] != self.EMPTY_CELL
                ):
                    return True
            # Check diagonals
            if (
                    self.board[0] == self.board[4] == self.board[8] != self.EMPTY_CELL
                    or self.board[2] == self.board[4] == self.board[6] != self.EMPTY_CELL
            ):
                return True
        return False

    def play_game(self):
        # Play a single game using the TicTacToeAgent
        self.reset()

        while True:
            # Get the current state
            state = self.get_state()
            # Let the agent choose an action
            action = self.agent.choose_action(state)
            # Make a move
            move_successful = self.make_move(action)
            self.print_board_after_move()  # Print the board after each move for debugging

            if not move_successful:
                continue  # If the move is invalid, try again

            # Check for a winner
            if self.check_winner():
                outcome = 'Win'
                print("Game Finished. Outcome: Win")
                break
            elif len([cell for cell in self.board if cell != self.EMPTY_CELL]) == 9:
                outcome = 'Draw'
                print("Game Finished. Outcome: Draw")
                break

        # Record game data
        moves = self.get_moves()
        self.games_data = self.games_data.append({'Moves': moves, 'Outcome': outcome}, ignore_index=True)

        # Print the final board state
        print("Final Board:")
        self.print_board()

        return outcome

    def get_state(self):
        # Return the current state of the board as input for the agent
        return np.array(self.board)

    def get_moves(self):
        # Return a list of moves made during the game
        return [i for i, cell in enumerate(self.board) if cell != self.EMPTY_CELL]

# Example usage:
num_episodes = 10
agent = TicTacToeAgent(num_actions=9)
env = TicTacToe(agent)

# Train the agent by playing multiple games
for episode in range(num_episodes):
    env.play_game()

# Access the games_data DataFrame after running the code
print("Games Data:")
print(env.games_data)
```
